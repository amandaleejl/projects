## Data Cleaning And Inspection for Weather Dataset

In this section, the weather dataset is imported, inspected and cleaned.

```python
# import libraries

# maths
import numpy as np
import pandas as pd

# others
import os
import re
import time
import datetime

# plotting
import matplotlib.pyplot as plt
import seaborn as sns

# import warnings filter
from warnings import simplefilter

# ignore all future warnings
simplefilter(action='ignore', category= (FutureWarning, UserWarning))
```

```python
# Load the weather dataset

df = pd.read_csv('./data/weather.csv')
```

```python
# Check if there are null values
# On 1st inspection there are 'null' values, but in the weather data dictionary missing values are input with 'M'

df.isnull().sum()
```

```python
# Change titles in columns to lower case.

df.columns = map(str.lower, df.columns)
```

```python
# Split Date (ojbect) into day, month and year
# This will be useful later when EDA and modelling are performed

def create_yr(x): 
    return x.split('-')[0] 

def create_mth(x): 
    return x.split('-')[1] 

def create_day(x): 
    return x.split('-')[2]

df['year'] = df.date.apply(create_yr)
df['month'] = df.date.apply(create_mth)
df['day'] = df.date.apply(create_day)
```

```python
# Confirm that the date is split

df.head()
```

### Inspection of the data indicated that there are 'M' ie missing values, Trace 'T', 0 and ' ' data.


```python
# For example, in the codesum column, there are ' ' empty columns.

df['codesum'].unique()
```

```python
# For example, snowfall has 0 values

df['snowfall'].unique()
```

```python
# Count total number of 'M', 'T''0' and ' ' in the df

def count_t(x): 
    if x == '  T':
        return 1
    else:
        return 0
    
def count_m(x): 
    if x == 'M':
        return 1
    else:
        return 0

def count_dash(x): 
    if x == '-':
        return 1
    else:
        return 0
    
def count_zero(x):
    if x == '0':
        return 1
    else:
        return
    
def count_zero2(x):
    if x =='0.0':
        return 1
    else:
        return

def count_space(x):
    if x==' ':
        return 1
    else:
        return

cols = ['column','M','-','T', '0', '0.0', ' ']
df_missing = pd.DataFrame(columns=cols)
i = 0

for col in df.columns:

    if df[col].dtype == 'object':

        total_m = df[col].apply(count_m).sum()
        total_dash = df[col].apply(count_dash).sum()
        total_t = df[col].apply(count_t).sum()
        total_z = df[col].apply(count_zero).sum()
        total_z2 = df[col].apply(count_zero2).sum()
        total_sp = df[col].apply(count_space).sum()
        
        df_missing.at[i,cols[0]] = col
        df_missing.at[i,cols[1]] = total_m
        df_missing.at[i,cols[2]] = total_dash
        df_missing.at[i,cols[3]] = total_t
        df_missing.at[i,cols[4]] = total_z
        df_missing.at[i,cols[5]] = total_z2
        df_missing.at[i,cols[6]] = total_sp
        i += 1
        
df_missing
```

### From  the data inspection the following can be performed :

Columns that can be removed :  depth, water1, snowfall.

codesum - ' ' impute with 'NS'

preciptotal - 'T' impute with '0' 

All data for sunrise and sunset in station 2 are empty. Impute data from station 1 since sunrise and sunset are almost the same within an area.

Impute tavg with (tmax-tmin)/2 + tmin

Impute depart with (tavg - depart) of station 1.

wetbulb, heat, cool, stnpressure, sealevel, avespeed - impute with median values or neighbouring values

```python
df.drop(['depth','water1', 'snowfall'], axis =1, inplace=True)
```

```python
df['codesum'].replace(to_replace=' ', value='NS', inplace=True)
```

```python
# Impute preciptotal missing value 'M' and trace 'T' with 0.0 since the data values are very small and insignifcant

df['preciptotal'].replace(to_replace='  T', value ='0.00', inplace=True)
df['preciptotal'].replace(to_replace='M', value ='0.00', inplace=True)
```

```python
# impute sunrise and sunset data from station 1
# station1 index is even number, station2 index is odd number
# sunrise is column 9, sunset is column 10

i = 0
while i < df.shape[0]:
    df.iloc[i+1, 9] = df.iloc[i, 9]
    df.iloc[i+1, 10] = df.iloc[i, 10]
    i+=2
```

```python
# impute tavg data with (tmax-tmin)/2

for i in range(df.shape[0]):
    if df['tavg'].at[i] =='M':
        df['tavg'].at[i] = (df['tmax'].at[i] - df['tmin'].at[i])*.05+df['tmin'].at[i]
```

```python
# impute depart with (tavg - depart) of station 1
# # station1 index is even number, station2 index is odd number

i = 0
for i in range(df.shape[0]-1):
    value = float(df['tavg'].at[i+1])-(float(df['tavg'].at[i])-float(df['depart'].at[i]))
    df['depart'].at[i+1] = str(value)
    i+=2 
```

```python
# impute wetbulb with the average values of preceding and proceding values

for i in range(df.shape[0]):
    if df['wetbulb'].at[i] =='M':
        df['wetbulb'].at[i] = str(float(df['wetbulb'].at[i+1]+df['wetbulb'].at[i-1])*.5)
```

```python
# impute heat and cool missing values 'M'
# The missing values are from station 2
df.loc[df['heat'] == 'M', 'station']
```

```python
for i in range(df.shape[0]):
    if df['heat'].at[i] =='M':
        df['heat'].at[i] = df['heat'].at[i-1]
    if df['cool'].at[i] == 'M':
        df['cool'].at[i] = df['cool'].at[i-1]
```

```python
# Impute stnpressure, sealevel, avgspeed with median values

for i in range(df.shape[0]):
    if df['stnpressure'].at[i] == 'M':
        df['stnpressure'].at[i] = df[df.stnpressure!='M'].stnpressure.median()
    if df['sealevel'].at[i] == 'M':
        df['sealevel'].at[i] = df[df.sealevel!='M'].sealevel.median()
    if df['avgspeed'].at[i] == 'M':
        df['avgspeed'].at[i] = df[df.avgspeed!='M'].avgspeed.median()
```

```python
# Distribution plot of continous variables

fig = plt.figure(figsize = (12,12))
ax = fig.gca()
df.hist(ax=ax)
plt.tight_layout()
```

```python
# Plot heatmap of the continous variables

plt.figure(figsize=(12,8))
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
ax = sns.heatmap(df.corr(), mask = mask, cmap = "coolwarm", annot = True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_ylim(bottom + 0.5, top - 0.5);
```
