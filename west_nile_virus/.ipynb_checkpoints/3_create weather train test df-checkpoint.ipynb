##  This section merges the cleaned weather dataset with the cleaned test and train datasets

```python
# Import libraries

import numpy as np 
import pandas as pd 
import time
import requests
import random
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'


pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
# Import the cleaned datasets

weather_clean = pd.read_csv("data/weather_cleaned.csv")
train_test = pd.read_csv("data/traintest.csv")            
```

```python
# Check the names of columns of the weather dataset

weather_clean.columns
```

```python
# Quick inspection of the weather and train-test datasets before merging

weather_clean.head()
```

```python
train_test.head(2)
```

```python
# Merge the datasets

df_merge = pd.merge(train_test, weather_clean, how="left", on=['date', 'station'] )
```

```python
# Inespect the merged datasets

df_merge.head()
```

```python
# Check the number of rows and columns of the merged datasaet

df_merge.shape
```

```python
# Save the merge dataset 

df_merge.to_csv('./data/weather_train_test.csv', index=False)
```
